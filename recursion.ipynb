##Tower of Hanoi Problem

image.png

optimum moves would be 2^n - 1 (n = number of disks)
arguments to consider ==> number of disks (n) , source , helper (auxiliary), destination

```python
def hanoi (disks, source, helper, destination):
  if(disks ==1):
    print('disk at {} , from {} tower, to {} tower'.format(disks, source, destination))

  else:
    hanoi(disks-1, source, destination, helper) #this one uses destination as helper
    print('disk at {} , from {} tower, to {} tower'.format(disks, source, destination))
    hanoi(disks-1, helper, source, destination) #this one uses source as helper

#changing order of the args keeps the recursion correctly
```

```python
disks = int(input("how many disks? "))
hanoi(disks, 'Source', 'Auxiliary', 'Destination')
```

##String Reversal Using Recursion

```python
#slice function
string = "fakeha"

new_string = string[::-1] #[starting point : ending point: step] since im not using any start/end leave it empty, and traversing from R-L
print(new_string)
```

```python
#Using Recursion
def reverse(str):
  if len(str) == 0:
    return str

  else:
    return reverse(str[1:]) + str[0]

string2 = "fakeha"
test_recursion = reverse(string2)
print(test_recursion)
```
